#Weather Data Cleansing

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pprint import pprint
from datetime import datetime
```

```python
Chicago_Weather = pd.read_csv("./Resources/NCDC_Weather_Data_Raw/Chicago_Weather.csv")
NYC_Weather = pd.read_csv("./Resources/NCDC_Weather_Data_Raw/NYC_Weather.csv")
DC_Weather = pd.read_csv("./Resources/NCDC_Weather_Data_Raw/DC_Weather.csv")
```

```python
NYC_Weather.head()
```

```python
Chicago_Weather = Chicago_Weather[['DATE', 'TAVG', 'TMAX', 'TMIN', 'PRCP', 'AWND', 'SNWD']]
Chicago_Weather["DATE"] = pd.to_datetime(Chicago_Weather["DATE"]).apply(lambda x: x.strftime('%m/%d/%Y'))

Chicago_Weather.head()
```

```python
NYC_Weather = NYC_Weather[['DATE', 'TAVG', 'TMAX', 'TMIN', 'PRCP', 'AWND', 'SNWD']]
NYC_Weather["DATE"] = pd.to_datetime(NYC_Weather["DATE"]).apply(lambda x: x.strftime('%m/%d/%Y'))

NYC_Weather.head()
```

```python
DC_Weather = DC_Weather[['DATE', 'TAVG', 'TMAX', 'TMIN', 'PRCP', 'AWND', 'SNWD']]
DC_Weather["DATE"] = pd.to_datetime(DC_Weather["DATE"]).apply(lambda x: x.strftime('%m/%d/%Y'))

DC_Weather.head()
```

```python
Chicago_Weather.to_csv("./Resources/Chicago_Weather_Clean.csv", index=False)
NYC_Weather.to_csv("./Resources/NYC_Weather_Clean.csv", index=False)
DC_Weather.to_csv("./Resources/DC_Weather_Clean.csv", index=False)
```
