```python
from splinter import Browser
from bs4 import BeautifulSoup
import time

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pprint import pprint
from datetime import datetime
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Load Chicago Divvy review webpage
Chicago_url = 'https://www.tripadvisor.com/Attraction_Review-g35805-d5074715-Reviews-Divvy_Bikes-Chicago_Illinois.html'
browser.visit(Chicago_url)
# HTML object
html = browser.html
# Parse HTML with Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')

# Retrieve page number
Chicago_review_data=[]
pagenumber = soup.find('div', class_='pageNumbers')
max_pagenumber = pagenumber.find('a', class_= 'pageNum last taLnk').text
max_pagenumber=int(max_pagenumber)
print(max_pagenumber)


# Retrieve each page of reviews
for page in range(max_pagenumber):
    print(page+1)
    href = 'https://www.tripadvisor.com/Attraction_Review-g35805-d5074715-Reviews-or' + str(page) + '0-Divvy_Bikes-Chicago_Illinois.html'
    print(href)
    
# Click "More" to show full text 
    browser.visit(href)
    browser.find_by_css('span.ulBlueLinks').click()
    time.sleep(1.001)
    print("clicked")

    # HTML object
    new_html = browser.html
    new_soup = BeautifulSoup(new_html, 'html.parser')
    
    valid_entries = new_soup.find_all('div', class_='rev_wrap ui_columns is-multiline')

    for entry in valid_entries:

        entry_userinfo = entry.find('div', class_='info_text')
        entry_username = entry_userinfo.find('div').text
        try: 
            entry_userloc = entry_userinfo.find('div', class_='userLoc').text
        except:
            entry_userloc = None
        stars = entry.find('span', class_='ui_bubble_rating')['class'][1]
        entry_date= entry.find('span', class_='ratingDate')['title']
        entry_quote= entry.find('div', class_='quote').text
        # entry_content= valid_entry.find('div', class_='prw_reviews_text_summary_hsx').text
        entry_content= entry.find('p', class_='partial_entry').text
        
        
        Chicago_review_data.append({"User_Name":entry_username,
                            "User_Location":entry_userloc,
                            "Stars":stars,
                            "Entry_Date":entry_date,
                            "Entry_Quote":entry_quote,
                            "Entry_Content":entry_content
                           })

        print('-----------')
        print(entry_username)
        print(entry_userloc)
        print(stars)
        print(entry_date)
        print(entry_quote)
        print(entry_content)
```

```python
Chicago_pd=pd.DataFrame()
for x in range(len(Chicago_review_data)):
    Chicago_pd.loc[x, 'User_Name'] = Chicago_review_data[x]['User_Name']
    Chicago_pd.loc[x, 'User_Location'] = Chicago_review_data[x]['User_Location']
    Chicago_pd.loc[x, 'Stars'] = Chicago_review_data[x]['Stars']
    Chicago_pd.loc[x, 'Entry_Date'] = datetime.strptime(Chicago_review_data[x]['Entry_Date'], '%B %d, %Y').strftime('%m/%d/%Y')
    
    Chicago_pd.loc[x, 'Entry_Quote'] = Chicago_review_data[x]['Entry_Quote']
    Chicago_pd.loc[x, 'Entry_Content'] = Chicago_review_data[x]['Entry_Content']

Chicago_pd.loc[Chicago_pd["Stars"] == "bubble_50", "Stars"] = 5
Chicago_pd.loc[Chicago_pd["Stars"] == "bubble_40", "Stars"] = 4
Chicago_pd.loc[Chicago_pd["Stars"] == "bubble_30", "Stars"] = 3
Chicago_pd.loc[Chicago_pd["Stars"] == "bubble_20", "Stars"] = 2
Chicago_pd.loc[Chicago_pd["Stars"] == "bubble_10", "Stars"] = 1

# datetime_object = datetime.strptime('April 15, 2019', '%B %d, %Y').strftime('%m/%d/%Y')

Chicago_pd.to_excel("./Resources/Web_Scraping_Data_Raw/Chicago_review_data.xlsx")
Chicago_pd.to_csv("./Resources/Web_Scraping_Data_Raw/Chicago_review_data.csv", index=False)
Chicago_pd.head()
```
