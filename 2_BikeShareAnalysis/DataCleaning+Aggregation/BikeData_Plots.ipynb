```python
chicago_2018["start_month"] = [datetime.strptime(x, "%m/%d/%Y").strftime("%m") for x in chicago_2018["start_date"]]
chicago_2018["start_month"] = pd.to_numeric(chicago_2018["start_month"])
chicago_2018.head()
```

```python
chicago_2018_groupby_month = chicago_2018.groupby(["start_month"])
chicago_2018_agg_month = chicago_2018_groupby_month.agg({"trip_count": "sum", "trip_duration_avg": "mean"})
chicago_2018_agg_month.reset_index(level=["start_month"], inplace=True)
chicago_2018_agg_month["trip_duration_avg"] = chicago_2018_agg_month["trip_duration_avg"].astype(int)

chicago_2018_agg_month
```
