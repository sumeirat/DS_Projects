```python
import numpy as np
import pandas as pd
```

```python
import datetime as dt
from datetime import datetime
import time
```

```python
# Weather Data for 2018
dc_weather = pd.read_csv("./Resources/Weather/DC_Weather_Clean.csv")
chicago_weather = pd.read_csv("./Resources/Weather/Chicago_Weather_Clean.csv")
```

```python
# Divvy Bikes
divvy_2014_Q1Q2 = pd.read_csv("./Resources/Bike_Data/Chicago_Divvy/2014/Divvy_Trips_2014_Q1Q2.csv")
divvy_2014_Q3_1 = pd.read_csv("./Resources/Bike_Data/Chicago_Divvy/2014/Divvy_Trips_2014-Q3-07.csv")
divvy_2014_Q3_2 = pd.read_csv("./Resources/Bike_Data/Chicago_Divvy/2014/Divvy_Trips_2014-Q3-0809.csv")
divvy_2014_Q4 = pd.read_csv("./Resources/Bike_Data/Chicago_Divvy/2014/Divvy_Trips_2014-Q4.csv")

divvy_1 = pd.read_csv("./Resources/Bike_Data/Chicago_Divvy/2018/Divvy_Trips_2018_Q1.csv")
divvy_2 = pd.read_csv("./Resources/Bike_Data/Chicago_Divvy/2018/Divvy_Trips_2018_Q2.csv")
divvy_3 = pd.read_csv("./Resources/Bike_Data/Chicago_Divvy/2018/Divvy_Trips_2018_Q3.csv")
divvy_4 = pd.read_csv("./Resources/Bike_Data/Chicago_Divvy/2018/Divvy_Trips_2018_Q4.csv")
```

```python
# Capital Bikes
capital_2011_all = pd.read_csv("./Resources/Bike_Data/DC_Capital/2011/2011-capitalbikeshare-tripdata.csv")
capital_2011 = pd.read_csv("./Resources/Bike_Data/DC_Capital/2011/2011-capitalbikeshare-tripdata.csv")

capital_18_1 = pd.read_csv("./Resources/Bike_Data/DC_Capital/2018/201801-capitalbikeshare-tripdata.csv")
capital_18_2 = pd.read_csv("./Resources/Bike_Data/DC_Capital/2018/201802-capitalbikeshare-tripdata.csv")
capital_18_3 = pd.read_csv("./Resources/Bike_Data/DC_Capital/2018/201803-capitalbikeshare-tripdata.csv")
capital_18_4 = pd.read_csv("./Resources/Bike_Data/DC_Capital/2018/201804-capitalbikeshare-tripdata.csv")
capital_18_5 = pd.read_csv("./Resources/Bike_Data/DC_Capital/2018/201805-capitalbikeshare-tripdata.csv")
capital_18_6 = pd.read_csv("./Resources/Bike_Data/DC_Capital/2018/201806-capitalbikeshare-tripdata.csv")
capital_18_7 = pd.read_csv("./Resources/Bike_Data/DC_Capital/2018/201807-capitalbikeshare-tripdata.csv")
capital_18_8 = pd.read_csv("./Resources/Bike_Data/DC_Capital/2018/201808-capitalbikeshare-tripdata.csv")
capital_18_9 = pd.read_csv("./Resources/Bike_Data/DC_Capital/2018/201809-capitalbikeshare-tripdata.csv")
capital_18_10 = pd.read_csv("./Resources/Bike_Data/DC_Capital/2018/201810-capitalbikeshare-tripdata.csv")
capital_18_11 = pd.read_csv("./Resources/Bike_Data/DC_Capital/2018/201811-capitalbikeshare-tripdata.csv")
capital_18_12 = pd.read_csv("./Resources/Bike_Data/DC_Capital/2018/201812-capitalbikeshare-tripdata.csv")
```

```python
divvy_1.columns = ['trip_id', 'start_time', 'end_time', 'bikeid', 'tripduration',
       'from_station_id', 'from_station_name', 'to_station_id',
       'to_station_name', 'usertype', 'gender', 'birthyear']
```

```python
divvy_dfs_2014 = [divvy_2014_Q1Q2, divvy_2014_Q3_1,  divvy_2014_Q3_2, divvy_2014_Q4]
divvy_dfs_2018 = [divvy_1, divvy_2, divvy_3, divvy_4]
divvy_2014 = pd.concat(divvy_dfs_2014, sort=False)
divvy_2018 = pd.concat(divvy_dfs_2018, sort=False)
```

```python
capital_dfs_2018 = [capital_18_1, capital_18_2, capital_18_3, capital_18_4, capital_18_5, capital_18_6,
               capital_18_7, capital_18_8, capital_18_9, capital_18_10, capital_18_11, capital_18_12]
capital_2018 = pd.concat(capital_dfs_2018)
```

```python
# ["trip_duration", "start_time", "end_time", "start_station_name", "end_station_name", "bikeid"]

divvy_2014.columns = ['trip_id', 'start_time', 'end_time', 'bikeid', 'trip_duration',
                      'start_station_id', 'start_station_name', 'end_station_id','end_station_name', 
                      'usertype', 'gender', 'birthyear']

divvy_2018.columns = ['trip_id', 'start_time', 'end_time', 'bikeid', 'trip_duration',
                      'start_station_id', 'start_station_name', 'end_station_id','end_station_name', 
                      'usertype', 'gender', 'birthyear']

capital_2011.columns = ['trip_duration', 'start_time', 'end_time', 'start_station_id',
       'start_station_name', 'end_station_id', 'end_station_name', 'bikeid', 'usertype']

capital_2018.columns = ['trip_duration', 'start_time', 'end_time', 'start_station_id',
       'start_station_name', 'end_station_id', 'end_station_name', 'bikeid', 'usertype']


divvy_2014 = divvy_2014[["trip_duration", "start_time", "end_time", "start_station_name", "end_station_name", "bikeid"]]
divvy_2018 = divvy_2018[["trip_duration", "start_time", "end_time", "start_station_name", "end_station_name", "bikeid"]]
capital_2011 = capital_2011[["trip_duration", "start_time", "end_time", "start_station_name", "end_station_name", "bikeid"]]
capital_2018 = capital_2018[["trip_duration", "start_time", "end_time", "start_station_name", "end_station_name", "bikeid"]]
```

```python
# Chicago Divvy - 2014
divvy_2014["start_time"] = [datetime.strptime(x, "%m/%d/%Y %H:%M") for x in divvy_2014["start_time"]]
divvy_2014["start_date"] = [x.strftime("%m/%d/%Y") for x in divvy_2014["start_time"]]
divvy_2014["start_hour"] = [x.strftime("%H") for x in divvy_2014["start_time"]]
divvy_2014["start_hour"] = pd.to_numeric(divvy_2014["start_hour"])
divvy_2014.head()
```

```python
# Chicago Divvy - 2018
divvy_2018["start_time"] = [datetime.strptime(x, "%Y-%m-%d %H:%M:%S") for x in divvy_2018["start_time"]]
divvy_2018["start_date"] = [x.strftime("%m/%d/%Y") for x in divvy_2018["start_time"]]
divvy_2018["start_hour"] = [x.strftime("%H") for x in divvy_2018["start_time"]]
divvy_2018["start_hour"] = pd.to_numeric(divvy_2018["start_hour"])
divvy_2018.head()
```

```python
# Chicago Divvy - 2018
divvy_2018["trip_duration"] = divvy_2018["trip_duration"].str.replace(',', '')
divvy_2018["trip_duration"] = pd.to_numeric(divvy_2018["trip_duration"])
divvy_2018.head()
```

```python
# DC Capital - 2011
capital_2011["start_time"] = [datetime.strptime(x, "%Y-%m-%d %H:%M:%S") for x in capital_2011["start_time"]]
capital_2011["start_date"] = [x.strftime("%m/%d/%Y") for x in capital_2011["start_time"]]
capital_2011["start_hour"] = [x.strftime("%H") for x in capital_2011["start_time"]]
capital_2011["start_hour"] = pd.to_numeric(capital_2011["start_hour"])
capital_2011.head()
```

```python
# DC Capital - 2018
capital_2018["start_time"] = [datetime.strptime(x, "%Y-%m-%d %H:%M:%S") for x in capital_2018["start_time"]]
capital_2018["start_date"] = [x.strftime("%m/%d/%Y") for x in capital_2018["start_time"]]
capital_2018["start_hour"] = [x.strftime("%H") for x in capital_2018["start_time"]]
capital_2018["start_hour"] = pd.to_numeric(capital_2018["start_hour"])
capital_2018.head()
```

```python
divvy_2014.drop(columns=["start_time", "end_time", "end_station_name"], inplace=True)
divvy_2018.drop(columns=["start_time", "end_time", "end_station_name"], inplace=True)

capital_2011.drop(columns=["start_time", "end_time", "end_station_name"], inplace=True)
capital_2018.drop(columns=["start_time", "end_time", "end_station_name"], inplace=True)
```

```python
divvy_2014_groupby = divvy_2014.groupby(["start_date", "start_hour", "start_station_name"])
divvy_2014_agg = divvy_2014_groupby.agg({"trip_duration": ["count", "mean"]})

divvy_2014_agg.columns = divvy_2014_agg.columns.droplevel()
divvy_2014_agg.rename(columns={"count": "trip_count", "mean": "trip_duration_avg"}, inplace=True)
divvy_2014_agg.reset_index(level=["start_date", "start_hour", "start_station_name"], inplace=True)
divvy_2014_agg["trip_duration_avg"] = divvy_2014_agg["trip_duration_avg"].astype(int)

divvy_2014_agg.head()
```

```python
divvy_2018_groupby = divvy_2018.groupby(["start_date", "start_hour", "start_station_name"])
divvy_2018_agg = divvy_2018_groupby.agg({"trip_duration": ["count", "mean"]})

divvy_2018_agg.columns = divvy_2018_agg.columns.droplevel()
divvy_2018_agg.rename(columns={"count": "trip_count", "mean": "trip_duration_avg"}, inplace=True)
divvy_2018_agg.reset_index(level=["start_date", "start_hour", "start_station_name"], inplace=True)
divvy_2018_agg["trip_duration_avg"] = divvy_2018_agg["trip_duration_avg"].astype(int)

divvy_2018_agg.head()
```

```python
capital_2011_groupby = capital_2011.groupby(["start_date", "start_hour", "start_station_name"])
capital_2011_agg = capital_2011_groupby.agg({"trip_duration": ["count", "mean"]})

capital_2011_agg.columns = capital_2011_agg.columns.droplevel()
capital_2011_agg.rename(columns={"count": "trip_count", "mean": "trip_duration_avg"}, inplace=True)
capital_2011_agg.reset_index(level=["start_date", "start_hour", "start_station_name"], inplace=True)
capital_2011_agg["trip_duration_avg"] = capital_2011_agg["trip_duration_avg"].astype(int)

capital_2011_agg.head()
```

```python
capital_2018_groupby = capital_2018.groupby(["start_date", "start_hour", "start_station_name"])
capital_2018_agg = capital_2018_groupby.agg({"trip_duration": ["count", "mean"]})

capital_2018_agg.columns = capital_2018_agg.columns.droplevel()
capital_2018_agg.rename(columns={"count": "trip_count", "mean": "trip_duration_avg"}, inplace=True)
capital_2018_agg.reset_index(level=["start_date", "start_hour", "start_station_name"], inplace=True)
capital_2018_agg["trip_duration_avg"] = capital_2018_agg["trip_duration_avg"].astype(int)

capital_2018_agg.head()
```

```python
chicago_merged_df = divvy_2018_agg.merge(chicago_weather, how="inner", left_on="start_date", right_on="DATE")
chicago_merged_df.drop(columns=["DATE"], inplace=True)
chicago_merged_df.head()
```

```python
dc_merged_df = capital_2018_agg.merge(dc_weather, how="inner", left_on="start_date", right_on="DATE")
dc_merged_df.drop(columns=["DATE"], inplace=True)
dc_merged_df.head()
```
