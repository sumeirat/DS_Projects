```python
import numpy as np
import pandas as pd
```

```python
# Reading aggregated bike datas for 2014(nyc&chicago) - 2011(dc)
chicago_2014 = pd.read_csv("./Resources/Bike_Data/Cleaned/chicago_agg_2014.csv")
dc_2011 = pd.read_csv("./Resources/Bike_Data/Cleaned/dc_agg_2011.csv")
nyc_2014 = pd.read_csv("./Resources/Bike_Data/Cleaned/nyc_agg_2014.csv")
```

```python
# Reading aggregated and merged bike datas for 2018 all cities
chicago_2018 = pd.read_csv("./Resources/Bike_Data/Cleaned/chicago_merged_2018.csv")
dc_2018 = pd.read_csv("./Resources/Bike_Data/Cleaned/dc_merged_2018.csv")
nyc_2018 = pd.read_csv("./Resources/Bike_Data/Cleaned/nyc_merged_2018.csv")
```

```python
# Reading Station Location Data
chicago_stations = pd.read_csv("./Resources/Bike_Data/Cleaned/Chicago_station_list.csv")
dc_stations = pd.read_csv("./Resources/Bike_Data/Cleaned/DC_station_list.csv")
nyc_stations = pd.read_csv("./Resources/Bike_Data/Cleaned/NYC_station_list.csv")
```

```python
chicago_stations.head()
```

```python
chicago_2018_gb = chicago_2018.groupby(["start_station_name"])
chicago_2018_agg = chicago_2018_gb.agg({"trip_count": "sum"}) 
chicago_2018_agg.reset_index(level=["start_station_name"], inplace=True)

chicago_2018_merged = chicago_2018_agg.merge(chicago_stations, how="inner", 
                                             left_on="start_station_name", right_on="Station Name")
chicago_2018_merged.drop(columns=["Station Name"], inplace=True)
chicago_2018_merged.rename(columns={"Latitude": "latitude", "Longitude": "longitude"}, inplace=True)
# chicago_2018_merged.sort_values(by='trip_count', ascending=False).head(50)

chicago_2018_merged.head()
```

```python
chicago_2014_gb = chicago_2014.groupby(["start_station_name"])
chicago_2014_agg = chicago_2014_gb.agg({"trip_count": "sum"}) 
chicago_2014_agg.reset_index(level=["start_station_name"], inplace=True)

chicago_2014_merged = chicago_2014_agg.merge(chicago_stations, how="inner", 
                                             left_on="start_station_name", right_on="Station Name")
chicago_2014_merged.drop(columns=["Station Name"], inplace=True)
chicago_2014_merged.rename(columns={"Latitude": "latitude", "Longitude": "longitude"}, inplace=True)

chicago_2014_merged.head()
```

```python
dc_stations.head()
```

```python
dc_2018_gb = dc_2018.groupby(["start_station_name"])
dc_2018_agg = dc_2018_gb.agg({"trip_count": "sum"}) 
dc_2018_agg.reset_index(level=["start_station_name"], inplace=True)

dc_2018_merged = dc_2018_agg.merge(dc_stations, how="inner", 
                                             left_on="start_station_name", right_on="ADDRESS")
dc_2018_merged.drop(columns=["ADDRESS"], inplace=True)
dc_2018_merged.rename(columns={"LATITUDE": "latitude", "LONGITUDE": "longitude"}, inplace=True)

dc_2018_merged.head()
```

```python
dc_2011_gb = dc_2011.groupby(["start_station_name"])
dc_2011_agg = dc_2011_gb.agg({"trip_count": "sum"}) 
dc_2011_agg.reset_index(level=["start_station_name"], inplace=True)

dc_2011_merged = dc_2011_agg.merge(dc_stations, how="inner", 
                                             left_on="start_station_name", right_on="ADDRESS")
dc_2011_merged.drop(columns=["ADDRESS"], inplace=True)
dc_2011_merged.rename(columns={"LATITUDE": "latitude", "LONGITUDE": "longitude"}, inplace=True)

dc_2011_merged.head()
```

```python
nyc_stations.head()
```

```python
nyc_2018_gb = nyc_2018.groupby(["start_station_name"])
nyc_2018_agg = nyc_2018_gb.agg({"trip_count": "sum"}) 
nyc_2018_agg.reset_index(level=["start_station_name"], inplace=True)

nyc_2018_merged = nyc_2018_agg.merge(nyc_stations, how="inner", 
                                             left_on="start_station_name", right_on="start station name")
nyc_2018_merged.drop(columns=["start station name", "index"], inplace=True)
nyc_2018_merged.rename(columns={"start station latitude": "latitude", "start station longitude": "longitude"}, inplace=True)

nyc_2018_merged.head()
```

```python
nyc_2014_gb = nyc_2014.groupby(["start_station_name"])
nyc_2014_agg = nyc_2014_gb.agg({"trip_count": "sum"}) 
nyc_2014_agg.reset_index(level=["start_station_name"], inplace=True)

nyc_2014_merged = nyc_2014_agg.merge(nyc_stations, how="inner", 
                                             left_on="start_station_name", right_on="start station name")
nyc_2014_merged.drop(columns=["start station name", "index"], inplace=True)
nyc_2014_merged.rename(columns={"start station latitude": "latitude", "start station longitude": "longitude"}, inplace=True)

nyc_2014_merged.head()
```

```python
map_firstyear_list = [chicago_2014_merged, dc_2011_merged, nyc_2014_merged]
map_firstyear = pd.concat(map_firstyear_list)
map_firstyear.count()
```

```python
map_2018_list = [chicago_2018_merged, dc_2018_merged, nyc_2018_merged]
map_2018 = pd.concat(map_2018_list)
map_2018.count()
```

```python
map_firstyear.to_csv("./Resources/Bike_Data/Cleaned/map_firstyear.csv", index=False)
map_2018.to_csv("./Resources/Bike_Data/Cleaned/map_2018.csv", index=False)
```
